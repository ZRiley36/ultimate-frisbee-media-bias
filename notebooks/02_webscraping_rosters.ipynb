```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

##Rosters

**In this notebook we will scrape the rosters of all the top 20 teams as ranked by USAU**.
[table](https://)

```python
response = requests.get("https://play.usaultimate.org/teams/events/team_rankings/?RankSet=College-Men")
soup = BeautifulSoup(response.content, "html.parser")
team_links = soup.find_all("tr")
```

Getting the links to each team roster from the rankings table on this [page](https://play.usaultimate.org/teams/events/team_rankings/?RankSet=College-Men):

```python
roster_links = []
for row in team_links:
  link = row.find('a')
  if link is not None:
    roster_links.append(link['href'])
```

```python
roster_links = roster_links[:20:]
# Here we change the last roster link to the 21st ranked team, as the 20th ranked team is d3
roster_links[19] = "/teams/events/Eventteam/?TeamId=ZGyVy1HcAUwCCpFKGwSblLfUyukqNpm%2fCQD%2busYlC3s%3d"
```

```python
import time
team_names = []
team_roster = []

for link in roster_links:
  link = 'https://play.usaultimate.org' + link
  response = requests.get(link)
  time.sleep(0.5)
  soup = BeautifulSoup(response.content, "html.parser")
  team_name = soup.find_all('h4')[0].text.strip()
  print(team_name)
  table = soup.find('table', attrs={'class':'global_table'})
  all_rows = table.find_all('tr')
  for row in all_rows:
    cols = row.find_all('td')
    if cols:
      player_name = cols[1].text
      team_names.append(team_name)
      team_roster.append(player_name)
```

populating a dataframe with the team name and players:

```python
df_rosters = pd.DataFrame()
df_rosters["Team"] = team_names
df_rosters["Player"] = team_roster
df_rosters
```

Adding a first and last names as seperate columns, for convienience, and saving the dataframe as a csv for later use:


```python
def get_first(x):
  names = x.split(' ')
  return names[0]


def get_last(x):
  names = x.split(' ')
  return names[1]

df_rosters["First"] = df_rosters["Player"].map(get_first)
df_rosters["Last"] = df_rosters["Player"].map(get_last)

df_rosters.to_csv("df_rosters.csv")
```
